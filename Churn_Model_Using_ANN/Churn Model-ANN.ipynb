# 1) Data Preprocessing

```python
import pandas as pd
df=pd.read_csv("Churn_Modelling.csv")
```

```python
df
```

### Row Number ,name and Customer ID are unnecessary data we won't use with. we have to develop a model which will predict whether customer is at risk of leaving by training on previous dataset.

```python
df=df.drop("RowNumber",axis=1)
df=df.drop("CustomerId",axis=1)
df=df.drop("Surname",axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

### There is no null data in our dataset , so we have no need to impute.

### Encoding

```python
import numpy as np
geo=df.iloc[:,1].values
```

```python
geo
```

```python
s=[]
unique=0
for i in geo:
    if(not (i in s)):
        s.append(i)
        unique+=1
print(unique)
```

### Total Number of Unique Countries is 3, we can perform OneHotEncoding on Geography, and LabelEncoding in Gender

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
enc=LabelEncoder()
df["Gender"]=enc.fit_transform(df["Gender"])
```

```python
x=df.iloc[:,:-1].values
df
```

```python
ct=ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[1])],remainder='passthrough')
x=ct.fit_transform(x)
x.shape
```

```python
y=df.iloc[:,-1:].values
x.shape
x
```

### We donot want high weights as it makes model unstable. it is always best to scale our data as it makes learning easier.


```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

### Data Preprocessing is complete.

# 2)Building an ANN 
we make 2 6 neuron hidden layer and a output layer, which predicts probability of customer leaving

```python
import tensorflow as tf
```

```python
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

```python
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

# 3)Training and analysis

```python
ann.fit(x_train,y_train,batch_size=20,epochs=100)
```

# Analysis with Test set:

```python
ypred=ann.predict(x_test)
```

```python
for i in range(0,2000):
    if(ypred[i]>0.5):
        ypred[i]=1
    else:
        ypred[i]=0
```

```python
from sklearn.metrics import classification_report,confusion_matrix
cm=confusion_matrix(y_test,ypred)
cr=classification_report(y_test,ypred)
print(cr)
```
